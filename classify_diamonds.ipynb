```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sb
```

```python
df = pd.read_csv('data/new_diamonds.csv')
df
```

```python
X = df.drop(['price', 'x', 'y', 'type'], axis=1)
X
```

```python
y = df['type'].to_frame()
y
```

```python
from sklearn.preprocessing import LabelEncoder
from pandas.api.types import is_integer_dtype, is_float_dtype, is_object_dtype
```

```python
X_new = pd.DataFrame()
for col in X:
    if is_object_dtype(X[col]):
        encoder = LabelEncoder()
        X_new[col] = encoder.fit_transform(X[col])
    else :
        X_new[col] = X[col]
```

```python
X_new
```

#### 피쳐 스케일링

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X_new)
X_scaled
```

#### 데이터 분할

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size = 0.2, random_state = 4)
```

#### KNN 통해 신규 데이터 분류

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
classifier.fit(X_train, y_train)
```

#### y 계산

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test, y_pred)
```
